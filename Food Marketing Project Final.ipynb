## Data Cleaning and Pre-processing

```python
# Checked the data for duplicates - Found 184 duplicates but we should have none due to how unlikely it is for nearly 40 columns to be identical
food_raw[food_raw.duplicated()].count() 
```

```python
# Dropped duplicates from the data
food_raw.drop_duplicates(keep = False, inplace = True)
```

```python
# Checked the dtypes of the data as well as how many null values are now present
food_raw.info()
```

```python
# Confirmed the removal of duplicates
food_raw[food_raw.duplicated()].count() 
```

###### Combined number of children per household into 1 column

```python
food_raw['Total_Children'] = food_raw[['Kidhome','Teenhome']].sum(axis=1) 
```

```python
# Replaced values in current cells with a number 1-5 for each marital status field
food_raw['marital_Divorced'] = food_raw['marital_Divorced'].replace({1:5,0:0})
food_raw['marital_Married'] = food_raw['marital_Married'].replace({1:4,0:0})
food_raw['marital_Single'] = food_raw['marital_Single'].replace({1:3,0:0})
food_raw['marital_Together'] = food_raw['marital_Together'].replace({1:2,0:0})
food_raw['marital_Widow'] = food_raw['marital_Widow'].replace({1:1,0:0})
```

```python
# Checked for that each row had a marital value between 1-5 
food_raw[food_raw['Marital_Status'] != 0].head()
```

###### Combined marital status values into one column 

```python
food_raw['Marital_Status'] = food_raw[['marital_Divorced','marital_Married','marital_Single','marital_Together','marital_Widow']].sum(axis=1) 
```

```python
# Checked that output worked as intended
food_raw.head()
```

###### Mapped marital status values to their contextual meanings

```python
food_raw['Marital_Status_str'] = food_raw['Marital_Status'].map({5: 'Divorced', 4: 'Married', 3: 'Single', 2: 'Together', 1: 'Widow'})
```

```python
# Checked that output worked as intended
food_raw.head()
```

```python
# Replaced values in current cells with a number 1-5 for each education field
food_raw['education_2n Cycle'] = food_raw['education_2n Cycle'].replace({1:1,0:0})
food_raw['education_Basic'] = food_raw['education_Basic'].replace({1:2,0:0})
food_raw['education_Graduation'] = food_raw['education_Graduation'].replace({1:3,0:0})
food_raw['education_Master'] = food_raw['education_Master'].replace({1:4,0:0})
food_raw['education_PhD'] = food_raw['education_PhD'].replace({1:5,0:0})
```

```python
# Checked that output worked as intended
food_raw.head()
```

```python
# Combined education level fields into one column 
food_raw['Education_Status'] = food_raw[['education_2n Cycle','education_Basic','education_Graduation','education_Master','education_PhD']].sum(axis=1) 
```

###### Mapped all education level values to their meanings

```python
food_raw['Education_Level'] = food_raw['Education_Level'].map({5: 'education_PhD', 4: 'education_Master', 3: 'education_Graduation', 2: 'education_Basic', 1: 'education_2n Cycle'})
```

```python
# Combined to one column 
food_raw['Accepted Campaigns'] = food_raw[['AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','AcceptedCmp1','AcceptedCmp2','Response']].sum(axis=1) 
```

```python
food_raw['Accepted Campaigns'] = (food_raw['Accepted Campaigns'] != 0).astype(int)
```

```python
# Checked for people who accepted campaigns to see if any accepted multiple campaigns (value > 1)
food_raw[food_raw['Accepted Campaigns'] != 0].head()
```

```python
# Segmented customers by age groups
age_groups = [(21,30),(31,40),(41,50),(51,60),(61,70),(71,95)]

def assign_age_group(Age):
    for age_range in age_groups:
        if age_range[0] <= Age <= age_range[1]:
            return f"{age_range[0]} - {age_range[1]}"
    return("Unknown")

food_raw['Age Group'] = food_raw['Age'].apply(assign_age_group)
```

```python
# Segmented customers by the total amount spent
total_spent = [(0,500),(501,1000),(1001,1500),(1501,2000),(2001,2500)]

def assign_spend_group(MntTotal):
    for amt_spent in total_spent:
        if amt_spent[0] <= MntTotal <= amt_spent[1]:
            return f"{amt_spent[0]} - {amt_spent[1]}"
    return("Unknown")

food_raw['Spending Groups'] = food_raw['MntTotal'].apply(assign_spend_group)
```
